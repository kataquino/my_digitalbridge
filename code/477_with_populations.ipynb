### Merge df's for population

```python
df = pd.merge(df_477, 
              df_pop[['census_tract_full', 'tract_only', 'population_tract', 'population_state']], 
              on='census_tract_full', how='left')
```

```python
# df["Order_Total"] = df.groupby('order')["ext price"].transform('sum')
```

```python
df[['population_tract', 'population_state']] = df[['population_tract', 'population_state']].astype(float)
```

```python
# state committment
df["pop_state_committed"] = df.groupby(['state_code', 'committed'])['population_tract'].transform('sum')
```

```python
# district committment
df["pop_district_committed"] = df.groupby(['LEAID', 'committed'])['population_tract'].transform('sum')
```

```python
# state tech committment
df['pop_state_tech_committed'] = df.groupby(['state_code', 'technology','committed'])['population_tract'].transform('sum')
```

```python
# district tech committment
df['pop_district_tech_committed'] = df.groupby(['LEAID', 'technology','committed'])['population_tract'].transform('sum')
```
